## Problem Statement
A retail company "ABC Private Limites" wants to understand the customer purchase behaviour (specially, purchase amount) against various products of different categories. They have shared purchase summery of various customers for selected high volume products from last month. The dataset also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details(product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of cutomer against the products which will help them to create personalized offer for customers against different products.

```python
#importing the dataset
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
#import test data
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
#Merge both train and test data
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
pd.get_dummies(df['Gender'],drop_first=1)
```

```python
##Handling categorical feature Gender- Convert categorical variable into numerical
#map method will map with respect to the condition that I am giving
print(df['Gender'].unique())
df.head()
```

```python
 df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
## Handle categorical feature Age
df['Age'].unique()
```

```python
#not work well
pd.get_dummies(df['Age'],drop_first=True).astype(int)
```

```python
#1st Technique
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7}) 
df.head()
```

```python
#Second Technique
from sklearn import preprocessing

#Label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

#Encode labels in columns 'Age'
df['Age']= label_encoder.fit_transform(df['Age'])

df['Age'].unique()
```
