```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
## To check Data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## Pie chart - Top3 counties that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%.2f%%')
```

Observation: Zomato maximum records or transactions are from India.
After that USA and then United Kingdom.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Rating are between 4.0 to 4.4 ---> very good
3. When Rating is between 3.5 to 3.9 ---> good
4. When Rating is between 3.0 to 3.4 ---> Average
5. When Rating is between 2.5 to 2.9 ---> Average
6. When Rating is between 2.0 to 2.4 ---> Average

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating", y= "Rating Count",data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y= "Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y= "Rating Count",hue='Rating color',data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y= "Rating Count",hue='Rating color',data=ratings,palette = ['white','red','orange','yellow','green','green'])
```

Observation:
1. Not Rated count is very high
2. Maximum number if rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x = "Rating color", data= ratings,hue='Rating color', palette = ['white','red','orange','yellow','green','green'])
```
